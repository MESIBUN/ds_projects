```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("obesity2.csv")    #Loading the dataset
df.head()
```

```python
df.shape #To see the number ofrows and columns
```

```python
df.describe() #Statistical ata for numeric columns
```

```python
df.columns # says about the columns present in the data set
```

```python
df.drop(columns ="TUE",inplace = True)
```

```python
df.info()
```

```python
df.rename(columns={'NObeyesdad': 'Obesity_status'}, inplace=True) #renaming the column
```

```python
df.rename(columns={'FAVC': 'FC_HCF'}, inplace=True) #FC_HCF - frequent consumption of high calories food
```

```python
df.rename(columns={'FCVC': 'FC_Veg'}, inplace=True) # Frequency of consumption of vegetables
```

```python
df.rename(columns={'NCP': 'No_MM'}, inplace=True)  #No.of main meals
```

```python
df.rename(columns={'CAEC': 'CF_BM'}, inplace=True) #Consumption of food meals
```

```python
df.rename(columns={'SCC': 'CCM'}, inplace=True) #Calories consumtion monitoring
```

```python
df.rename(columns={'MTRANS': 'T_used'}, inplace=True) #Transportation used
```

```python
df.isnull().sum() #checking null values
```

```python
# Data preprocessing

#Encoding categorical to numerical using LabelEncoding

from sklearn.preprocessing import LabelEncoder
columns_to_encode = ["Gender","family_history_with_overweight","FC_HCF","CF_BM","SMOKE","CCM","CALC","T_used"]
le = LabelEncoder()
df[columns_to_encode] = df[columns_to_encode].apply(lambda col: le.fit_transform(col))
```

```python
df.info()
```

```python
df.head()
```

```python
df["Obesity_status"]=le.fit_transform(df["Obesity_status"])
```

# Obesity status:


0- Insufficient Weight

1- Normal Weight

2- Overweight Level I

3- Overweight Level II

4- Obesity Type I

5- Obesity Type II

6- Obesity Type III






```python
df["Obesity_status"].value_counts()
```

```python
X = df.drop(columns="Obesity_status")
X.shape
```

```python
y = df["Obesity_status"]
y.shape
```

```python
# Building model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print("X_train :",X_train.shape)
print("X_test :",X_test.shape)
print("y_train :",y_train.shape)
print("y_test :",y_test.shape)
```

```python
# Standardizing  the features (optional but recommended for logistic regression)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Standardizing  the features (optional but recommended for logistic regression)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Creating a logistic regression model
model = LogisticRegression()
# Train the model
model.fit(X_train, y_train)
```

```python
# Making predictions on the test set
predictions = model.predict(X_test)
```

```python
 #Evaluating the model
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)
```

```python
# Printing the results
print("Accuracy:", accuracy)
print("Classification Report:\n", report)
```

```python
# Decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.tree import plot_tree


# Create a decision tree classifier
clf = DecisionTreeClassifier()

# Fit the classifier to the training data
clf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = clf.predict(X_test)

# Evaluate the accuracy of the model
accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# Support vector Machines(svm)
from sklearn.svm import SVC
# Creating an SVM classifier
svm_classifier = SVC(kernel='linear', C=1.0)

# Training the SVM classifier
svm_classifier.fit(X_train, y_train)

# Making predictions on the test set
y_pred = svm_classifier.predict(X_test)

# Evaluating the accuracy of the model

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

```

```python
y_pred = svm_classifier.predict([[0,23,1.66,68.0,1,0,2.0,2,1,1,2.0,	1,0.0,2,3]])
print(y_pred)
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
# Creating a Random Forest classifier
random_forest_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Training the Random Forest classifier
random_forest_classifier.fit(X_train, y_train)

# Making predictions on the test set
y_pred = random_forest_classifier.predict(X_test)

# Evaluating the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
